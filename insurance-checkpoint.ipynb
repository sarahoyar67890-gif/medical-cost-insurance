```python
df.replace({'sex': {'male':0, 'female':1},'smoker': {'yes':0, 'no':1},'region': {'southeast':0,'southwest':1,'northeast':2,'northwest':3 }}, inplace=True)
```

```python
df
```

```python
#spliting features and target
x=df.drop(columns='charges',axis=1)
y=df['charges']
```

```python
x
```

```python
y
```

```python
#spliting the data into training and testing 
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
#model training
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
#model evalution
#predicting on training data
y_pred_train=regressor.predict(x_train)
```

```python
r_square_train_value = metrics.r2_score(y_train, y_pred_train)
print("R squared value:", r_square_train_value)
```

```python
#predicting on testing data
y_pred_test=regressor.predict(x_test)
```

```python
r_square_test_value = metrics.r2_score(y_test, y_pred_test)
print("R squared value:", r_square_test_value)
```
